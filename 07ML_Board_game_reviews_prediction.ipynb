```python
#Remove games without users ratings
gamedata = gamedata[gamedata['average_rating'] > 0]

#Drop missing values
gamedata = gamedata.dropna(axis=0)

#Visualize using histogram
plt.hist(gamedata['average_rating'])
plt.show()
```

```python
#Visualizing the correlation matrix

cormat = gamedata.corr()
fig = plt.figure(figsize = (20, 10))
sns.heatmap(cormat, vmax=.8, annot = True)
plt.show()
```

```python
#get all the columns in the dataframe
columns = gamedata.columns.tolist()

#filter out unwanted columns
columns = [c for c in columns if c not in ['bayes_average_rating', 'average_rating', 'type', 'name', 'id']]

#store the data we will be using

target = "average_rating"
```

```python
#generate the training and test datasets

from sklearn.model_selection import train_test_split
train = gamedata.sample(frac=0.8, random_state=1)
#items not in the trainig should be in the test

test = gamedata.loc[~gamedata.index.isin(train.index)]

print(train.shape)
print(test.shape)
```

```python
#Import Linear Regreesion model
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

LR = LinearRegression()

#Fiting the model
LR.fit(train[columns], train[target])

#Making Prediction
predictions = LR.predict(test[columns])

#Calculating the meansquare error
mean_squared_error(predictions, test[target])
```

```python
#Trying out the Random Forest Model

from sklearn.ensemble import RandomForestRegressor

RFR = RandomForestRegressor(n_estimators = 100, min_samples_leaf=10, random_state=1)

#fitting the model

RFR.fit(train[columns], train[target])

#Making predictions

RFRPredict = RFR.predict(test[columns])

#Checking mean squared error

mean_squared_error(RFRPredict, test[target])
```

```python
#Predictng for  Individaul games

LR_ratings = LR.predict(test[columns].iloc[0].values.reshape(1, -1))
RFR_ratings = RFR.predict(test[columns].iloc[0].values.reshape(1, -1))


#Print Prediction

print(LR_ratings)
print(RFR_ratings)
```
